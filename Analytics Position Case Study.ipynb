```python
## Calculation of deposit points
## formula = weightage per activity (0.01)*deposit amount
```

```python
deposit_amt_pts = 0.01*deposit['Amount'].fillna(0)
```

```python
deposit_amt_pts
```

```python
## withdrawal amount game points calculation
## Formula = Weightage per activity (0.005)*Wiwithdrawal_amt
```

```python
withdrawal_amt_pts = 0.005*withdrawal['Amount'].fillna(0)
```

```python
withdrawal_amt_pts
```

```python
##How many more times did a player do deposit than withdrawal?

## Formula = 0.001*MAX(deposit-withdrawal,0)

##0.001 * max((5-3, 0)) = 0.001 * 2 
## = 0.002 points

## where number of deposit = 5
## and number of withdrawal = 3
```

```python
deposit_count = deposit['User Id'].value_counts().rename('deposit_count')
withdrawal_count = withdrawal['User Id'].value_counts().rename('withdrawal_count')
```

```python
transaction_count = pd.concat([deposit_count, withdrawal_count], axis=1).fillna(0)
```

```python
transaction_count
```

```python
## difference of deposit count and withdrawal count
transaction_count['moredeposits_thanwithdrawal'] = (transaction_count['deposit_count']-transaction_count['withdrawal_count']).astype(int)
```

```python
transaction_pts = transaction_count['moredeposits_thanwithdrawal'].apply(
    lambda x: 0.001 * max(x, 0)
).fillna(0)
```

```python
## points calculation
## 0.001 * max((5-3, 0)) = 0.001 * 2
```

```python
## CALCULATION OF NO. OF games played
## Formula= 0.2*Number of Games played 
```

```python
games_pts = 0.2*game['Games Played'].fillna(0)
```

```python
## CALCULATION OF LOYALTY POINTS
## Loyalty_pts = deposit_amt_pts+withdrawal_amt_pts+transaction_pts+games_pts
```

```python
game['deposit_amt_pts'] = deposit_amt_pts.fillna(0)
game['withdrawal_amt_pts'] = withdrawal_amt_pts.fillna(0)
game['transaction_pts'] = transaction_pts.fillna(0).fillna(0)
game['games_pts'] = games_pts.fillna(0)
```

```python
game['loyalty_pts'] = (game['deposit_amt_pts'].fillna(0) +
    game['withdrawal_amt_pts'].fillna(0) +
    game['transaction_pts'].fillna(0) +
    game['games_pts'].fillna(0))
```

```python
game['deposit_amt_pts'] = game['deposit_amt_pts'].fillna(0)
game['withdrawal_amt_pts'] = game['withdrawal_amt_pts'].fillna(0)
game['transaction_pts'] = game['transaction_pts'].fillna(0)
game['games_pts'] = game['games_pts'].fillna(0)
game['loyalty_pts'] = game['loyalty_pts'].fillna(0)
```
